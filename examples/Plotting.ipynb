# Plotting

```python
import datetime
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from iexfinance import get_historical_data

plt.style.use("fivethirtyeight")

start = datetime.datetime(2017,1,1)
end = datetime.datetime(2018,1,1)
```

The simplest way to obtain a stock chart is by simply plotting the ``close`` column of the OHLC DataFrame.

```python
df = get_historical_data("AAPL", start, end, output_format='pandas')["close"]
df.index = pd.DatetimeIndex(df.index)
df.plot(title="AAPL 2017 Chart")
```

The same can be done for multiple symbols:

```python
from pandas_datareader import data as web

syms = ["AAPL", "TSLA", "DAL"]
df = web.DataReader(syms, "iex",start,end)
df["close"].plot()
```

## Moving Averages

```python
df = pd.concat([df["close"],df["close"].rolling(window=20).mean()], axis=1)
```

```python
df.plot()
```
